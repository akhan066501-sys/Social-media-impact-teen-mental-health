## Step 5: Feature Engineering
Creating two new categorical columns to enable group-based analysis:
- **Age Group** — Categorizing teenagers into age brackets (13-15, 16-17, 18+)
- **Media Usage Category** — Classifying daily social media hours as Low, Medium, or High

```python
# Creating Age Group column
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
bins = [12, 15, 17, 200]
labels = ['13-15', '16-17', '18+']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=True, include_lowest=True)
data['Age Group'] = data['Age Group'].cat.add_categories(['Unknown']).fillna('Unknown')
data['Age Group'].value_counts()
```

```python
# Creating Media Usage Category column
# Low = 0-3 hrs/day, Medium = 3-6 hrs/day, High = 6-10 hrs/day
data['Media Usage Category'] = pd.cut(data['Daily Social Media Hours'], 
                                       bins=[0, 3, 6, 10], 
                                       labels=['Low', 'Medium', 'High'])
data['Media Usage Category'].value_counts()
```

## Step 6: Group-wise Analysis
Grouping the data by Age Group, Gender, and Platform to understand how different segments behave in terms of social media usage, sleep, and mental health metrics.

```python
# Multiple aggregated statistics grouped by Age Group
age_group_stats = data.groupby('Age Group').agg({
    'Daily Social Media Hours': ['mean', 'min', 'max'],
    'Sleep Hours': ['mean', 'std'],
    'Stress Level': 'mean',
    'Anxiety Level': 'mean',
    'Depression Label': 'sum'
})
age_group_stats
```

```python
# Filtering high social media users — more than 6 hours daily
high_users = data[data['Daily Social Media Hours'] > 6]
print(f"Number of high users: {len(high_users)}")
high_users.head()
```

```python
# Average social media hours, sleep, and academic performance grouped by Gender
data.groupby('Gender')[['Daily Social Media Hours', 'Sleep Hours', 'Academic Performance']].mean()
```

```python
# Distribution of platform usage across the dataset
data['Platform Usage'].value_counts()
```

```python
# Gender distribution in the dataset
data['Gender'].value_counts()
```

## Step 7: Detailed Pandas Analysis
Building detailed summary tables using groupby and pivot operations to extract meaningful insights from the data.

```python
# Comprehensive summary statistics by Age Group
age_group_summary = data.groupby('Age Group').agg({
    'Age': 'count',
    'Daily Social Media Hours': 'mean',
    'Sleep Hours': 'mean',
    'Academic Performance': 'mean',
    'Stress Level': 'mean',
    'Anxiety Level': 'mean',
    'Depression Label': 'sum'
}).round(2)
age_group_summary.columns = ['Count', 'Avg Daily Media Hours', 'Avg Sleep', 'Avg Academic Performance', 'Avg Stress', 'Avg Anxiety', 'Depression Cases']
age_group_summary
```

```python
# Mental health metrics analyzed by Media Usage Category
mental_health_by_usage = data.groupby('Media Usage Category')[['Stress Level', 'Anxiety Level', 'Depression Label']].agg(['mean', 'median', 'std'])
mental_health_by_usage
```

```python
# Pivot table — Average sleep hours by Platform and Gender
pivot_sleep = data.pivot_table(values='Sleep Hours', index='Platform Usage', columns='Gender', aggfunc='mean')
pivot_sleep
```

## Step 8: Correlation Analysis
Measuring the statistical relationship between variables — particularly between social media usage and mental health indicators.

- **Positive correlation (+1):** Both variables increase together
- **Negative correlation (-1):** One increases while the other decreases
- **Zero (0):** No relationship between the variables

```python
# Full correlation matrix for all numerical columns
correlation_matrix = data.corr(numeric_only=True)
correlation_matrix
```

```python
# Correlation specifically between mental health and social media usage
mental_health_cols = ['Daily Social Media Hours', 'Stress Level', 'Anxiety Level', 'Depression Label']
mental_health_corr = data[mental_health_cols].corr()
mental_health_corr
```

```python
# Factors most correlated with Depression Label (sorted descending)
depression_corr = data.corr(numeric_only=True)['Depression Label'].sort_values(ascending=False)
print("Factors correlated with Depression:")
depression_corr
```

## Step 9: Data Visualization
Visual representation of patterns, distributions, and relationships in the data using matplotlib and seaborn.

```python
# Heatmap showing correlation between all variables
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix - All Variables')
plt.tight_layout()
plt.show()
```

```python
# 4-panel chart — Distribution, Gender comparison, Social Media vs Stress, Social Media vs Sleep
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histogram — distribution of daily social media hours
data['Daily Social Media Hours'].hist(ax=axes[0,0], bins=30, edgecolor='black')
axes[0,0].set_title('Distribution of Daily Social Media Hours')
axes[0,0].set_xlabel('Hours')

# Box plot — social media hours by gender
data.boxplot(column='Daily Social Media Hours', by='Gender', ax=axes[0,1])
axes[0,1].set_title('Social Media Hours by Gender')

# Scatter plot — Social Media Hours vs Stress Level
axes[1,0].scatter(data['Daily Social Media Hours'], data['Stress Level'], alpha=0.5)
axes[1,0].set_xlabel('Daily Social Media Hours')
axes[1,0].set_ylabel('Stress Level')
axes[1,0].set_title('Social Media Hours vs Stress Level')

# Scatter plot — Social Media Hours vs Sleep Hours
axes[1,1].scatter(data['Daily Social Media Hours'], data['Sleep Hours'], alpha=0.5, color='green')
axes[1,1].set_xlabel('Daily Social Media Hours')
axes[1,1].set_ylabel('Sleep Hours')
axes[1,1].set_title('Social Media Hours vs Sleep Hours')

plt.tight_layout()
plt.show()
```

## Step 10: At-Risk Teenager Identification
Identifying teenagers who exhibit both high stress and high anxiety simultaneously — this group is considered at high risk for mental health issues.



```python
# Filter teenagers with high stress AND high anxiety
at_risk = data[(data['Stress Level'] > 6) & (data['Anxiety Level'] > 6)]
print(f"Number of at-risk teenagers: {len(at_risk)}")
print(f"Percentage of total: {(len(at_risk)/len(data)*100):.2f}%")
at_risk[['Age', 'Gender', 'Daily Social Media Hours', 'Stress Level', 'Anxiety Level', 'Depression Label']].head(10)
```

### Conclusions
The analysis clearly demonstrates that **excessive social media usage has a measurable negative impact** on teenager mental health:
- **Sleep quality decreases** as social media hours increase
- **Stress and anxiety levels rise** with higher usage
- **Academic performance declines** with more screen time
- A significant percentage of teenagers fall into the **at-risk category** based on combined stress and anxiety scores

### Recommendations
> - Teenagers should limit daily social media usage to **under 3 hours**
> - **Avoid screen time at least 1 hour before sleep** to improve sleep quality
> - Parents and educators should monitor teenagers showing high addiction levels
> - Mental health awareness programs should target the **13–15 age group** which shows the highest vulnerability
