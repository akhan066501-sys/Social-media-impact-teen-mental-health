#  Social Media Impact on Teenager Mental Health

**Objective:** Analyze how social media usage affects the mental health of teenagers — including stress, anxiety, sleep, and academic performance.

**Dataset:** Raw_Data.csv — 1200 records of teenagers aged 13–19

**Columns:** Age, Gender, Daily Social Media Hours, Platform Usage, Sleep Hours, Screen Time Before Sleep, Academic Performance, Physical Activity, Social Interaction Level, Stress Level, Anxiety Level, Addiction Level, Depression Label

Sabse pehle zaroori libraries import karte hain — data manipulation, visualization aur analysis ke liye.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

CSV file ko pandas DataFrame mein load karte hain.

```python
data = pd.read_csv('Raw_Data.csv')
```

## ⚙️ Step 5: Feature Engineering
Analysis ke liye do naye columns banate hain:
- **Age Group** — Age ko categories mein divide karna (13-15, 16-17, 18+)
- **Media Usage Category** — Social media hours ko Low / Medium / High mein classify karna

```python
# Age Group column banana
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
bins = [12, 15, 17, 200]
labels = ['13-15', '16-17', '18+']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=True, include_lowest=True)
data['Age Group'] = data['Age Group'].cat.add_categories(['Unknown']).fillna('Unknown')
data['Age Group'].value_counts()
```

```python
# Media Usage Category column banana — Low (0-3 hrs), Medium (3-6 hrs), High (6-10 hrs)
data['Media Usage Category'] = pd.cut(data['Daily Social Media Hours'], 
                                       bins=[0, 3, 6, 10], 
                                       labels=['Low', 'Medium', 'High'])
data['Media Usage Category'].value_counts()
```

## 📊 Step 6: Group-wise Analysis
Age Group, Gender aur Platform ke hisaab se data ko group karke average values nikalte hain.

```python
# Age Group ke hisaab se multiple statistics
age_group_stats = data.groupby('Age Group').agg({
    'Daily Social Media Hours': ['mean', 'min', 'max'],
    'Sleep Hours': ['mean', 'std'],
    'Stress Level': 'mean',
    'Anxiety Level': 'mean',
    'Depression Label': 'sum'
})
age_group_stats
```

```python
# High social media users filter karna (>6 hours daily)
high_users = data[data['Daily Social Media Hours'] > 6]
print(f"Number of high users: {len(high_users)}")
high_users.head()
```

```python
# Gender ke hisaab se average social media, sleep aur academic performance
data.groupby('Gender')[['Daily Social Media Hours', 'Sleep Hours', 'Academic Performance']].mean()
```

```python
# Platform usage ka distribution
data['Platform Usage'].value_counts()
```

```python
# Gender distribution
data['Gender'].value_counts()
```

## 🐼 Step 7: Pandas Data Analysis
Age Group aur Media Usage Category ke basis pe detailed summary tables banate hain.

```python
# Age Group ke hisaab se summary statistics
age_group_summary = data.groupby('Age Group').agg({
    'Age': 'count',
    'Daily Social Media Hours': 'mean',
    'Sleep Hours': 'mean',
    'Academic Performance': 'mean',
    'Stress Level': 'mean',
    'Anxiety Level': 'mean',
    'Depression Label': 'sum'
}).round(2)
age_group_summary.columns = ['Count', 'Avg Daily Media Hours', 'Avg Sleep', 'Avg Academic Performance', 'Avg Stress', 'Avg Anxiety', 'Depression Cases']
age_group_summary
```

```python
# Media Usage Category ke hisaab se mental health analysis
mental_health_by_usage = data.groupby('Media Usage Category')[['Stress Level', 'Anxiety Level', 'Depression Label']].agg(['mean', 'median', 'std'])
mental_health_by_usage
```

```python
# Platform aur Gender ke hisaab se sleep hours ka pivot table
pivot_sleep = data.pivot_table(values='Sleep Hours', index='Platform Usage', columns='Gender', aggfunc='mean')
pivot_sleep
```

## 🔗 Step 8: Correlation Analysis
Dekhte hain ki kaun se variables ek doosre se related hain — khaas taur pe social media aur mental health ke beech ka connection.

```python
# Poore dataset ki correlation matrix
correlation_matrix = data.corr(numeric_only=True)
correlation_matrix
```

```python
# Mental health aur social media usage ke beech correlation
mental_health_cols = ['Daily Social Media Hours', 'Stress Level', 'Anxiety Level', 'Depression Label']
mental_health_corr = data[mental_health_cols].corr()
mental_health_corr
```

```python
# Depression Label se sabse zyada correlated factors
depression_corr = data.corr(numeric_only=True)['Depression Label'].sort_values(ascending=False)
print("Factors correlated with Depression:")
depression_corr
```

## 📈 Step 9: Data Visualization
Graphs aur charts ke through data ko visually samajhte hain — distribution, patterns aur relationships.

```python
# Correlation Heatmap — saare variables ke beech relationship
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix - All Variables')
plt.tight_layout()
plt.show()
```

```python
# 4 charts — Distribution, Gender comparison, Social Media vs Stress, Social Media vs Sleep
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histogram — daily social media hours ka distribution
data['Daily Social Media Hours'].hist(ax=axes[0,0], bins=30, edgecolor='black')
axes[0,0].set_title('Distribution of Daily Social Media Hours')
axes[0,0].set_xlabel('Hours')

# Box plot — Gender ke hisaab se social media hours
data.boxplot(column='Daily Social Media Hours', by='Gender', ax=axes[0,1])
axes[0,1].set_title('Social Media Hours by Gender')

# Scatter plot — Social Media vs Stress Level
axes[1,0].scatter(data['Daily Social Media Hours'], data['Stress Level'], alpha=0.5)
axes[1,0].set_xlabel('Daily Social Media Hours')
axes[1,0].set_ylabel('Stress Level')
axes[1,0].set_title('Social Media Hours vs Stress Level')

# Scatter plot — Social Media vs Sleep Hours
axes[1,1].scatter(data['Daily Social Media Hours'], data['Sleep Hours'], alpha=0.5, color='green')
axes[1,1].set_xlabel('Daily Social Media Hours')
axes[1,1].set_ylabel('Sleep Hours')
axes[1,1].set_title('Social Media Hours vs Sleep Hours')

plt.tight_layout()
plt.show()
```

## ⚠️ Step 10: At-Risk Teenagers Identification
Un teenagers ko identify karte hain jinka **Stress Level > 6 AND Anxiety Level > 6** hai — ye high-risk group hai.

```python
# High stress AND high anxiety wale teenagers
at_risk = data[(data['Stress Level'] > 6) & (data['Anxiety Level'] > 6)]
print(f"Number of at-risk teenagers: {len(at_risk)}")
print(f"Percentage of total: {(len(at_risk)/len(data)*100):.2f}%")
at_risk[['Age', 'Gender', 'Daily Social Media Hours', 'Stress Level', 'Anxiety Level', 'Depression Label']].head(10)
```

- **Academic performance pe bura asar** padta hai
